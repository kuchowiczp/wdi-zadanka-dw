Wykorzystując bibliotekę `pandas`, załaduj plik `iris.csv` do ramki danych (dataframe). Wyświetl pojedyncze punkty ze zbioru danych na 4 wykresach (każdy z parametrów z każdym) punktowych. Punkty powinny mieć różne kolory w zależności od gatunku irysa (class). Wykorzystaj funkcję `plot.scatter()`.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("iris.csv")

color_map = {
    'Iris-setosa': 'lawngreen',
    'Iris-versicolor': 'tomato',
    'Iris-virginica': 'navy'
}
df['color'] = df['class'].map(color_map)

print(df.columns)
```

```python
df.plot.scatter(x='sepallength', y='sepalwidth', c=df['color'])
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
df.plot.scatter(x='petallength', y='petalwidth', c=df['color'])
plt.title('Petal Length vs Petal Width')
plt.show()
```

```python
df.plot.scatter(x='sepallength', y='petallength', c=df['color'])
plt.title('Sepal Length vs Petal Length')
plt.show()
```

```python
df.plot.scatter(x='sepalwidth', y='petalwidth', c=df['color'])
plt.title('Sepal Width vs Petal Width')
plt.show()
```
